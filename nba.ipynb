```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
nba_data = pd.read_excel("NBA_Stats_71_Years_Updated.xlsx")
nba_data.head(3)
```

```python
nba_data.dtypes
```

```python
nba_data.head(3)
```

```python
nba_data.isnull().sum()
```

```python
rows = len(nba_data)
percentofnull = ((nba_data.isnull().sum())/rows)*100
percentofnull
```

```python
x = nba_data["3-Pt FG Made"]
fig = px.histogram(nba_data, x)

fig.show()
```

```python
nba_data['3-Pt FG Attempts'] = nba_data['3-Pt FG Attempts'].fillna(nba_data['3-Pt FG Attempts'].median())
nba_data['3-Pt FG Made'] = nba_data['3-Pt FG Made'].fillna(nba_data['3-Pt FG Made'].median())
```

```python
nba_data.isnull().sum()
```

```python
threepointscolumns = [ "Year", "Season Start Year", "Season Type", "Player", "Team","Games Played","3-Pt FG Attempts", "3-Pt FG Made" ]
threepointsdf = nba_data[threepointscolumns]
threepointsdf = threepointsdf.reset_index()
threepointsdf = threepointsdf.drop("index", axis=1)
threepointsdf
```

```python
threepointsdf = threepointsdf.loc[threepointsdf["Season Type"] == "Playoffs"]
```

```python
threepointsdf.isnull().sum()
```

```python
three_for_graph = threepointsdf.groupby("Season Start Year").sum()
three_for_graph
```

```python
x = three_for_graph.index
a = three_for_graph["Games Played"]
b = three_for_graph["3-Pt FG Attempts"]
c = three_for_graph["3-Pt FG Made"]

# Create traces
fig = go.Figure()
fig.add_trace(go.Bar(x=x, y=a, name="Games Played"))
fig.add_trace(go.Scatter(x=x, y=b, mode="markers+lines", name="3 PTS Attempt"))
fig.add_trace(go.Scatter(x=x, y=c, mode="markers+lines", name="3 PTS Made"))


fig.show()
```

```python
who_shot_three = threepointsdf.groupby("Player").sum()
who_shot_three
```

```python
who_shot_three = who_shot_three.sort_values("3-Pt FG Made", ascending=False)
best_players_who_shot_three = who_shot_three[:10]
best_players_who_shot_three = best_players_who_shot_three.drop(axis=1,columns="Season Start Year")
best_players_who_shot_three
```

```python
best_players = best_players_who_shot_three.index
best_players
```

```python
# Create traces for best 3 points players


games_played = best_players_who_shot_three["Games Played"]
pts_attempts = best_players_who_shot_three["3-Pt FG Attempts"]
pts_made = best_players_who_shot_three["3-Pt FG Made"]
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=best_players, y=games_played, text=games_played, name="Games Played"))
fig.add_trace(go.Scatter(x=best_players, y=pts_attempts, mode="markers+lines", name="3 PTS Attempt"))
fig.add_trace(go.Scatter(x=best_players, y=pts_made, mode="markers+lines", name="3 PTS Made"))

fig.show()
```

```python
threepointsdf = threepointsdf[threepointsdf["Player"].isin(best_players)]
threepointsdf
```

```python
x = threepointsdf["3-Pt FG Attempts"]
y = threepointsdf["3-Pt FG Made"]
fig = px.scatter(threepointsdf, x, y)
fig.show()
```

```python
x = threepointsdf["Season Start Year"]
y = threepointsdf["3-Pt FG Made"]

fig = px.scatter(threepointsdf, x, y, color="Player", size="Games Played",animation_group="Player")

#fig.update_layout(height=700)
fig.show()
```
